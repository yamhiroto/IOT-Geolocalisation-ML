Dcomposition spectrale de $\Sigma_n$

Analyse des valeurs propres (spectre)

```python
X_train.shape
```

```python
X_tilde = np.dot(X_train, eig_vec_60.T) # La nouvelle matrice X  considrer
```

```python
X_tilde.shape
```

```python
from sklearn.linear_model import LinearRegression
reg_pca = LinearRegression(fit_intercept=True).fit(X_tilde, Y_train)
```

```python
xxx = np.arange(0, len(reg_pca.coef_[0]))
```
