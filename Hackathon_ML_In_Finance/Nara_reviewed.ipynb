```python
import pandas as pd
import numpy as np
from sklearn import tree
import os
import time
```

```python
file_address = os.getcwd() + "/data/"
train_explanatory = pd.read_csv(file_address + 'sp-explanatory-train.csv')
train_response = pd.read_csv(file_address + 'sp-response-train.csv')

test_explanatory = pd.read_csv(file_address + 'sp-explanatory-valid.csv', header=None)
test_explanatory.columns = train_explanatory.columns.tolist()
```

```python
log_returns_train = pd.DataFrame( columns=train_explanatory.columns.tolist()[3:])
log_returns_test = pd.DataFrame( columns=test_explanatory.columns.tolist()[3:])
for i in range(train_explanatory.shape[0]):
    log_returns_train = log_returns_train.append((train_explanatory.iloc[i,2:].map(np.log) - 
                                                  train_explanatory.iloc[i,2:].shift(1).map(np.log))[1:])

for i in range(test_explanatory.shape[0]):
    log_returns_test = log_returns_test.append((test_explanatory.iloc[i,2:].map(np.log) - 
                                                test_explanatory.iloc[i,2:].shift(1).map(np.log))[1:])
```

### New dataframe definition

```python
df_sp = train_explanatory.drop(['openGammaCall','openGammaPut'], axis=1)
df_sp
```

### Vectorization (1)

```python
start_time = time.time()
for i in range(train_explanatory.shape[0]):
    log_returns_train = log_returns_train.append((train_explanatory.iloc[i,2:].map(np.log) - 
                                                  train_explanatory.iloc[i,2:].shift(1).map(np.log))[1:])
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
start_time = time.time()
df_sp_ret = df_sp.T.apply(lambda x: (x/ x.shift(1, axis='rows')) - 1)
df_sp_ret_transpose = df_sp_ret.T
df_sp_ret_final = df_sp_ret_transpose.drop(df_sp_ret_transpose.columns[[0]],axis=1)
print("--- %s seconds ---" % (time.time() - start_time))
```

### What is autocorrelation?

```python
s = pd.Series([0.25, 0.5, 0.2, -0.05])
```

```python
s.autocorr(lag=2)
```

```python
a = pd.Series([0.25, 0.5])
b = pd.Series([0.2, -0.05])
```

```python
a.corr(b)
```

#### covariance

```python
cov = 1/2*((0.25 - np.mean(a))*(0.2 - np.mean(b)) + (0.5 - np.mean(a))*(-0.05 - np.mean(b)))
```

```python
cov
```

####variance

```python
((0.25-0.375)**2+(0.5-0.375)**2)/2.0
```

```python
np.var(a)
```

```python
np.std(a)
```

#### Pearson coef

```python
cov/(np.std(a)*np.std(b))
```

```python
start_time = time.time()

diff_gamma = train_explanatory.openGammaCall - train_explanatory.openGammaPut
list_diff_gamma = diff_gamma.tolist()
ret_spx_4h = []
list_var_spx_log_ret =[]
list_var_spx= []
autocorr_lag10_log_ret =[]

for i in range(train_explanatory.shape[0]):
    ret_spx_4h.append(train_explanatory.iloc[i,-1] - train_explanatory.iloc[i,2]) # rendements absolus premier VS dernier
    list_var_spx_log_ret.append(log_returns_train.iloc[i].var()) # variance des rendements relatifs
    autocorr_lag10_log_ret.append(log_returns_train.iloc[i].autocorr(lag=10)) # autocorrelation des rendements dcals
    # autocorr: Pearson correlation between vect1 and vect2.shift(lag)
    # vect2.shift(3):
    # a = pd.Series([1, 2, 3, 4, 5, 6])
    # a.shift(3)
    # output:= [nan, nan, nan, 1, 2, 3]
    # a.autocorr(lag=3) is equivalent to [4,5,6].corr([1,2,3])
    # in other words, autocorr(lag=10) does the correlation between 2 vectors: one without the last 10 values, one without the first 10 values
X=[]
for i in range(train_explanatory.shape[0]):
    
    X.append([
              list_diff_gamma[i],
              ret_spx_4h[i],
              list_var_spx_log_ret[i],
              autocorr_lag10_log_ret[i]
             ])  
Y = train_response.Y.tolist()

print("--- %s seconds ---" % (time.time() - start_time))
```

### Vectorization (2)

```python
start_time = time.time()

x_gamma = train_explanatory.openGammaCall - train_explanatory.openGammaPut
x_gamma = x_gamma.to_frame()
x_gamma.columns = ['gammaDiff']

x_last = df_sp[df_sp.columns[[df_sp.shape[1]-1]]].apply(lambda x: x - np.ones(df_sp.shape[0])*100)
x_last.columns = ['s&pDiff']

x_var_rel = df_sp_ret_final.T.apply(lambda x: np.var(x)).to_frame()
x_var_rel.columns = ['s&pVar']

x_autocorr = df_sp_ret_final.T.apply(lambda x: x.autocorr(lag=10)).to_frame()
x_autocorr.columns = ['s&pAutocorr']

X = pd.concat([x_gamma, x_last, x_var_rel, x_autocorr], axis=1)

Y = train_response.Y.tolist()

print("--- %s seconds ---" % (time.time() - start_time))
```

```python
diff_gamma_test = test_explanatory.openGammaCall - test_explanatory.openGammaPut
list_diff_gamma_test = diff_gamma_test.tolist()
ret_spx_4h_test = []
list_var_spx_log_ret_test =[]
list_var_spx_test =[]
autocorr_lag10_log_ret_test =[]

for i in range(test_explanatory.shape[0]):
    ret_spx_4h_test.append(test_explanatory.iloc[i,-1] - test_explanatory.iloc[i,2])
    list_var_spx_log_ret_test.append(log_returns_test.iloc[i].var())
    list_var_spx_test.append(test_explanatory.iloc[i][2:].var())
    autocorr_lag10_log_ret_test.append(log_returns_test.iloc[i].autocorr(lag=10))
X_test=[]
for i in range(test_explanatory.shape[0]):
    
    X_test.append([list_diff_gamma_test[i],
                   ret_spx_4h_test[i],
                   list_var_spx_log_ret_test[i],
                   autocorr_lag10_log_ret_test[i]
             ]) 
```

```python
clf = tree.DecisionTreeClassifier(random_state = 0)
clf = clf.fit(X,Y)
Df_rest = pd.DataFrame (clf.predict(X_test), columns= ['Y'])
#Df_rest.to_csv('test.txt',  index = False, header =False)
#Df_rest.to_csv('sp-team-nara.csv',  index = False, header =False)
```
