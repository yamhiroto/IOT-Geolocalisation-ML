```python
# 1.
import pandas as pd
data = pd.read_csv("https://bitbucket.org/portierf/shared_files/downloads/Galton.txt", sep='\t')
```

```python
data
```

```python
# 2.
data["MeanParents"] = 0.5*(data["Father"] + 1.08*data["Mother"])
```

```python
data
```

```python
# 3.
import matplotlib.pyplot as plt

x = data["MeanParents"]
y = data["Height"]

fig = plt.figure(figsize=(12,12))
plt.scatter(x,y)
plt.xlabel('Taille parents')
plt.ylabel('Taille enfants') 
plt.grid(True)
plt.show()
```

On voit que la taille des enfants est linéaire avec la taille des parents

```python
#4.
import numpy as np
from sklearn.linear_model import LinearRegression
x = np.array(x).reshape(-1,1)
# Note: sans reshape, np.array(x) renvoie un vecteur de nombres --> .fit() requiert un vecteur d'array
y = np.array(y).reshape(-1,1)

# fit:
# trouve les coefficients optimaux (la méthode d'optimisation utilisée n'est pas l'inversion)
# = étape d'entraînement
# on utilise 'fit_intercept' car les données ne sont pas centrées et on veut la constante
reg = LinearRegression(fit_intercept=True).fit(x, y)

# predict:
# utilise les coefficients estimés pour calculer Y
y_pred_1 = reg.predict(x)

fig = plt.figure(figsize=(12,12))
plt.scatter(x,y)
plt.scatter(x,y_pred_1)
plt.xlabel('Taille parents')
plt.ylabel('Taille enfants')
plt.show()
```

<h1><center><i>fit()</i> (cas multivarié)</center></h1>
$$(X^TX)\widehat{\theta}=X^TY$$

<h1><center><i>predict()</i> (cas multivarié)</center></h1>
$$pred(X)=X\widehat{\theta}$$

La régression calculée donne une droite qui passe par le milieu du nuage de points.

```python
# 5.
# Recentrage: on retranche la moyenne de toutes les observations pour la taille des parents et celle des enfants
y_c = y - y.mean()
x_c = x - x.mean()

# On calcule les coefficients de la régression dans le cas centré
reg_c = LinearRegression(fit_intercept=False).fit(x_c, y_c)

y_pred_c = reg_c.predict(x_c)

# On compare les résultats obtenus avec les résultats précédents à l'aide de la formule (cf plus bas)

# Comparaison des coefficients
# Cas non centré
theta_0 = reg.intercept_
theta_1 = reg.coef_
# Cas centré
theta_0_c = y.mean() - x.mean()*reg_c.coef_
theta_1_c = reg_c.coef_

print("Dans le cas non centré, theta0 = %f" % theta_0[0])
print("Dans le cas centré, theta0 = %f" % theta_0_c[0][0])
print("")
print("Dans le cas non centré, theta1 = %f" % theta_1[0][0])
print("Dans le cas centré, theta1 = %f" % theta_1_c[0][0])

# Comparaison des prédictions
y_pred_centre=reg_c.predict(x_c)
np.allclose(y_pred_centre+y.mean(),y_pred_1)
```

On voit qu'on récupère bien les mêmes résultats d'optimisation dans le cas centré et non centré grâce à la formule vue en cours:

$$\widehat{\theta}=
\quad
\begin{pmatrix}
\overline{Y}-\overline{\tilde{X^T}}\tilde{\theta} \\
\tilde{\theta}
\end{pmatrix}
\quad
$$

```python
# 6.
res = y - y_pred_1
plt.hist(res)
plt.title('Résidus', fontsize=10)
plt.show()
```

$$r_{i}=y_{i}-\widehat{y_{i}}$$

L'hypothèse de normalité ne semble pas très crédible au vu de la forme de la courbe (la courbe n'est pas très en cloche).

On peut aussi calculer le kurtosis pour mieux décrire la forme de la courbe.
Ce kurtosis étant très faible (négatif), cela signifie que la distribution est bien plus aplatie que celle de la loi normale:

```python
from scipy.stats import kurtosis
print(kurtosis(res))
```

```python
# 7.
x_filles = np.array(data[data["Gender"] == "F"]["MeanParents"]).reshape(-1,1)
y_filles = np.array(data[data["Gender"] == "F"]["Height"]).reshape(-1,1)

x_garcons = np.array(data[data["Gender"] == "M"]["MeanParents"]).reshape(-1,1)
y_garcons = np.array(data[data["Gender"] == "M"]["Height"]).reshape(-1,1)


fig = plt.figure(figsize=(12,12))
plt.scatter(x,y_pred_1)
plt.scatter(x_filles,y_filles,c='coral')
plt.scatter(x_garcons,y_garcons,c='lightblue')
plt.xlabel('Taille parents')
plt.ylabel('Taille enfants')
plt.show()
```

On voit qu'il y a de fortes disparités entre la taille des garçons et la taille des filles.
Ces différences créent des écarts importants entre les valeurs réelles et celles prédites, d'où la forme épaisse de la courbe des résidus.

```python
# 8.
reg_garcons = LinearRegression(fit_intercept=True).fit(x_garcons, y_garcons)
y_pred_garcons_1 = reg_garcons.predict(x_garcons)

fig1 = plt.figure(figsize=(12,12))
plt.scatter(x,y_pred_1)
plt.scatter(x_filles,y_filles,c='coral')
plt.scatter(x_garcons,y_garcons,c='lightblue')
plt.scatter(x_garcons,y_pred_garcons_1,c='orange')
plt.xlabel('Taille parents')
plt.ylabel('Taille enfants')
plt.show()

res_garcons = y_garcons - y_pred_garcons_1
fig2 = plt.figure(figsize=(12,12))
plt.hist(res_garcons)
plt.title('Résidus régression garçons', fontsize=10)
plt.show()
```

On remarque que la distribution des résidus ressemble davantage à une courbe gaussienne lorsque l'on prend des individus garçons uniquement. On peut vérifier cette information en calculant le kurtosis, plus élevé qué précédemment:

```python
print(kurtosis(res_garcons))
```

```python
# 9.
from sklearn.model_selection import train_test_split

data_train, data_test = train_test_split(data, test_size=0.2)

# Echantillon de test
x_test_garcons = np.array(data_test[data_test["Gender"]=="M"]["MeanParents"]).reshape(-1,1)
y_test_garcons = np.array(data_test[data_test["Gender"]=="M"]["Height"]).reshape(-1,1)

# Echantillon d'apprentissage
# Modèle 1 - garçons uniquement
x_train_garcons = np.array(data_train[data_train["Gender"] == "M"]["MeanParents"]).reshape(-1,1)
y_train_garcons = np.array(data_train[data_train["Gender"] == "M"]["Height"]).reshape(-1,1)
reg_train_garcons = LinearRegression(fit_intercept=True).fit(x_train_garcons, y_train_garcons)
y_pred_garcons_2 = reg_train_garcons.predict(x_test_garcons)
# Modèle 2 - échantillon complet
x_train_all = np.array(data_train["MeanParents"]).reshape(-1,1)
y_train_all = np.array(data_train["Height"]).reshape(-1,1)
reg_train_all = LinearRegression(fit_intercept=True).fit(x_train_all, y_train_all)
y_pred_2 = reg_train_all.predict(x_test_garcons)

N = len(y_test_garcons)

rPred_garcons = 0
for i in range(N):
    rPred_garcons+=((y_test_garcons[i] - y_pred_garcons_2[i])**2)
rPred_garcons = rPred_garcons/N

rPred_all = 0
for i in range(len(x_test_garcons)):
    rPred_all+=((y_test_garcons[i] - y_pred_2[i])**2)
rPred_all = rPred_all/N

print("Le risque de prédiction du modèle basé sur les garçons est %f" % rPred_garcons)
print("Le risque de prédiction du modèle basé sur l'échantillon complet est %f" % rPred_all)
```

On constate que le risque prédictif est plus faible dans le cas d'une prédiction sur les garçons seulement. Ce résultat est attendu car, comme vu à la question 8, les erreurs résiduels sont plus faibles lorsque le calcul des coefficients est fait sur l'échantillon des garçons uniquement.

```python
# 10.
x_test_filles = np.array(data_test[data_test["Gender"]=="F"]["MeanParents"]).reshape(-1,1)
y_test_filles = np.array(data_test[data_test["Gender"]=="F"]["Height"]).reshape(-1,1)

# On rajoute un modèle: Modèle 3 - filles uniquement
x_train_filles = np.array(data_train[data_train["Gender"] == "F"]["MeanParents"]).reshape(-1,1)
y_train_filles = np.array(data_train[data_train["Gender"] == "F"]["Height"]).reshape(-1,1)
reg_train_filles = LinearRegression(fit_intercept=True).fit(x_train_filles, y_train_filles)
y_pred_filles = reg_train_filles.predict(x_test_filles)

# On combine les résultats des tests filles / garçons
y_test = np.concatenate([y_test_garcons, y_test_filles])
y_pred_3 = np.concatenate([y_pred_garcons_2, y_pred_filles])

# On calcule le risque prédictif total
N = len(y_test)
rPred_combined = 0
for i in range(N):
    rPred_combined+=((y_test[i] - y_pred_3[i])**2)
rPred_combined = rPred_combined/N

print("Le risque de prédiction sur l'échantillon complet " +
      "avec modèle différent selon fille/garçon est %f" % rPred_combined)
```

On constate que le risque prédictif total prenant compte du modèle fille/garçon selon le genre est bien inférieur au risque de prédiction basé sur l'échantillon complet qui ne fait pas la distinction fille/garçon.
